```python
def draw_line(coef,intercept, mi, ma):
    # for the separating hyper plane ax+by+c=0, the weights are [a, b] and the intercept is c
    # to draw the hyper plane we are creating two points
    # 1. ((b*min-c)/a, min) i.e ax+by+c=0 ==> ax = (-by-c) ==> x = (-by-c)/a here in place of y we are keeping the minimum value of y
    # 2. ((b*max-c)/a, max) i.e ax+by+c=0 ==> ax = (-by-c) ==> x = (-by-c)/a here in place of y we are keeping the maximum value of y
    points=np.array([[((-coef[1]*mi - intercept)/coef[0]), mi],[((-coef[1]*ma - intercept)/coef[0]), ma]])
    plt.plot(points[:,0], points[:,1])
```

```python
def Linear_models(ratios, C, classifier):

    for j,i in enumerate(ratios):
        plt.figure(figsize=(20,5))
        
        for k in range(len(C)):
            plt.subplot(1, 3, k+1)
            np.random.seed(0)
            X_p=np.random.normal(0,0.05,size=(i[0],2))
            X_n=np.random.normal(0.13,0.02,size=(i[1],2))
            y_p=np.array([1]*i[0]).reshape(-1,1)
            y_n=np.array([0]*i[1]).reshape(-1,1)
            X=np.vstack((X_p,X_n))
            y=np.vstack((y_p,y_n))
            
            classifier.C = C[k]
            classifier.fit(X,y)
            
            coef, intercept = classifier.coef_[0], classifier.intercept_
            mi, ma = X.min(), X.max()
            draw_line(coef,intercept, mi, ma)

            plt.scatter(X_p[:,0],X_p[:,1])
            plt.scatter(X_n[:,0],X_n[:,1],color='red')
            plt.title('C = {}, Ratio = {}'.format(C[k], i))
    plt.show()
```

```python
ratios = [(100,2), (100, 20), (100, 40), (100, 80)]
C = [0.001, 1, 100]

res = Linear_models(ratios,C, SVC(kernel = 'linear' ))
res
```

### Observations for Linear SVM

- Based on the above plots, we can see that, the learning rate(C) impacts the hyperplane which seperates the blue points from red points

When C = 0.001 : 
- As the C is very small, model does not classify the data for both balanced as well as imbalanced dataset.We can see that the hyperplane is far away from datapoints. 
- Here the value of C is small and hence the loss term is very small and negligible as compared to the regularization term. So the model tends to underfit and therefore the margin increases such that most of the datapoints would get misclassified.

When C = 1 :
- When the data is extremely imbalanced(100,2), highly imbalanced(100,20) and imbalanced (100,40) then we can see that it does not classify the points but in almost balanced data, the model classifies the datapoints with some errors.

When C = 100 :
- It does not works for extremely imbalanced dataset even if the value of C is high. But for the remaining dataset it works fine
- When the value of C is large, the loss term keeps dominating the regularization term and finally the regularization term becomes negligible when compared to loss term. Therefore the model tends to overfit and the margin tends to reduces such that most of the datapoints are classified correctly.
- But it does not garrenty that, this model performs well on future unseen data.
